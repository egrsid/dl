### Обработка датасета

Была проведена выше с объяснениями. Здесь будет объединение всего в несколько функций для удобного вызова и обработки датасетов

```python
def change_y_n(data, cat_cols):
  for feature in cat_cols:
    feat = data[feature]
    if isinstance(feat.unique()[0], str):
      if len(feat.unique()) == 2 and feat.unique()[0].lower() in ['yes', 'no']:
        data[feature].replace(feat.unique(), [1, 0], inplace=True)
  return data
```

```python
def change_sex_and_totalSpent(data):
  data['Sex'].replace(['Male', 'Female'], [0, 1], inplace=True)
  data.TotalSpent = data.TotalSpent.replace([' '], [0]).astype(float)
  return data
```

```python
def change_starting_with_no(data, features):
  for feat in features:
    f = [x for x in data[feat].unique() if x.startswith('No ')]
    data[feat].replace(['Yes', 'No', f[0]], [1, 0, 0], inplace=True)
  return data
```

```python
def encode_features(encoder, data, cat_feat):
  one_hot_encoded = encoder.fit_transform(data[cat_feat])
  one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cat_feat), index=data.index)

  df = data.drop(cat_feat, axis=1)
  df = pd.concat([df, one_hot_df], axis=1)

  return df
```

```python
def prepare_dataset(data, cat_cols, cat_feat, features, encoder):
  data = change_y_n(data.copy(), cat_cols)
  data = change_sex_and_totalSpent(data.copy())
  data = change_starting_with_no(data.copy(), features)
  result = encode_features(encoder, data.copy(), cat_feat)

  return result
```

```python
def oversampling(data, target):
  sampler = ADASYN()
  x_resampled, y_resampled = sampler.fit_resample(data.drop([target], axis=1), data[target])

  return x_resampled, y_resampled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import optuna
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
import warnings
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_resampled, y_resampled, shuffle=True, stratify=y_resampled, test_size=0.2)
```

```python
def objective(trial, model_name):
    if model_name == 'LogisticRegressionCV':
        model = LogisticRegressionCV(
            Cs = trial.suggest_int('Cs', 1, 10),
            cv=trial.suggest_int('cv', 3, 10),
            max_iter = trial.suggest_int('max_iter', 1000, 2000),
            scoring='roc_auc',
            penalty='l2',
            n_jobs=-1
        )

    elif model_name == 'LogisticRegression':
        model = LogisticRegression(
            max_iter = trial.suggest_int('max_iter', 1000, 2000),
            C = trial.suggest_categorical('C', [0.1, 10, 0.01]),
            solver = trial.suggest_categorical('solver', ['newton-cholesky', 'liblinear']),
            penalty='l2',
            n_jobs=-1
        )

    model.fit(x_train, y_train)
    y_score = model.predict_proba(x_test)[:, 1]
    return roc_auc_score(y_test, y_score)
```

```python
models = [LogisticRegression(n_jobs=-1), LogisticRegressionCV(n_jobs=-1)]

optuna.logging.set_verbosity(optuna.logging.INFO)

result = {}
with warnings.catch_warnings(record=False):
  for model in models:
      print(f"Оптимизация модели: {model.__class__.__name__}")
      study = optuna.create_study(direction="maximize", study_name=f"{model.__class__.__name__} optimal parameter search")
      study.optimize(lambda trial: objective(trial, model.__class__.__name__), n_trials=50)

      result[model.__class__.__name__] = study.best_params
      print("-" * 50, '\n')
```

Модель на модифицированном сете справляется лучше. Среди бустингов и логрегов лучше будет бустинг. То есть беру его как конечную модель для предсказания. Теперь нужно подбирать для него параметры

```python
def objective(trial):
    model = CatBoostClassifier(
            iterations=trial.suggest_int('iterations', 1000, 2000),
            depth=trial.suggest_int('depth', 3, 10),
            learning_rate=trial.suggest_float('learning_rate', 3e-2, 1e-1, log=True),
            early_stopping_rounds=20,  # борьба против переобучения
            silent=True  # чтобы не выводился лосс по итерациям
        )

    model.fit(x_train, y_train)
    y_score = model.predict_proba(x_test)[:, 1]
    return roc_auc_score(y_test, y_score)
```

```python
study = optuna.create_study(direction="maximize", study_name=f"CatBoost optimal parameter search")
study.optimize(objective, n_trials=50)
```

```python
params = study.best_params
params
```

```python
boosting = CatBoostClassifier(silent=True)
boosting.set_params(**params)
boosting.fit(x_train, y_train)
```

```python
print(f'Roc_auc_score: {roc_auc_score(y_test, boosting.predict_proba(x_test)[:, 1])}')
```
